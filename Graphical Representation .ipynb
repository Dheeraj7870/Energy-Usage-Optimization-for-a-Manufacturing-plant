Heatmap

```python
numerical_columns = df.select_dtypes(exclude=['object']).columns

numerical_df = df[numerical_columns]

print(numerical_df)
```

```python
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 6))  
sns.heatmap(correlation_matrix, annot=True, cmap='YlGn', fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap for Numerical Data", fontsize=16)
plt.show()
```

1. Relationships:

. Melting Time and Cooling Energy: There is a strong positive correlation between these two variables. This suggests that longer melting times are associated with higher cooling energy consumption.

. Annealing Time and Cooling Energy: There is a moderate positive correlation between these two variables. This suggests that longer annealing times might be associated with slightly higher cooling energy consumption.

. Recycled Content and Defects Percentage: There is a moderate positive correlation between these two variables. This might indicate that higher recycled content could be associated with a slightly higher defect percentage.

Other Observations:

. There are several moderate to weak correlations between other variables, but they don't seem to have strong implications.

. Most of the correlations are relatively weak, suggesting that the relationships between variables might be complex and influenced by multiple factors.

2.Potential Business Implications:

. Energy Efficiency: Focusing on reducing melting and annealing times could lead to significant energy savings.

. Quality Control: Understanding the relationship between recycled content and defects could help improve quality control measures.

. Process Optimization: Further analysis is needed to identify other potential areas for process optimization and energy efficiencyimprovements.
